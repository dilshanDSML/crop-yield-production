# Scaling and Onehotencoding

```python
col=["Year","average_rain_fall_mm_per_year","pesticides_tonnes","avg_temp","Area","Item","hg/ha_yield"]
df=df[col]
```

```python
df.head()
```

```python
X=df.drop("hg/ha_yield",axis=1)
y=df["hg/ha_yield"]
X.head()
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,               # Features and labels
    test_size=0.33,     # 25% for testing, 75% for training
    random_state=42,    # Reproducible results
            # Shuffle before splitting
)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
ohe=OneHotEncoder(drop="first")
scaler=StandardScaler()
```

```python
preprocessor=ColumnTransformer(transformers=[("ohe1",ohe,[4,5]),("standardization",scaler,[0,1,2,3])],remainder="passthrough")
```

```python
preprocessor
```

```python
X_train_dummy=preprocessor.fit_transform(X_train)
X_test_dummy=preprocessor.transform(X_test)
```

# Training Model

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
models={
    "lr":LinearRegression(),
    "lss":Lasso(),
    "rdg":Ridge(),
    "knr":KNeighborsRegressor(),
    "dtr":DecisionTreeRegressor()
}
for name,mod in models.items():
    mod.fit(X_train_dummy,y_train)
    y_pred=mod.predict(X_test_dummy)
    print(f"{name}  Score : {r2_score(y_test,y_pred)}")
    
```

# Select Model

```python
dtr=DecisionTreeRegressor()
dtr.fit(X_train_dummy,y_train)
dtr.predict(X_test_dummy)
```

# Prediction System

```python
def prediction(Year,average_rain_fall_mm_per_year,pesticides_tonnes,avg_temp,Area,Item):
    features=np.array([[Year,average_rain_fall_mm_per_year,pesticides_tonnes,avg_temp,Area,Item]])
    transformed_feature=preprocessor.transform(features)
    predicted_value=dtr.predict(transformed_feature).reshape(1,-1)
    return predicted_value
```

```python
Year=2000
average_rain_fall_mm_per_year=59.0
pesticides_tonnes=3024.11
avg_temp=26.55	
Area="Saudi Arabia"	
Item="Sorghum"
prediction(Year,average_rain_fall_mm_per_year,pesticides_tonnes,avg_temp,Area,Item)
```
